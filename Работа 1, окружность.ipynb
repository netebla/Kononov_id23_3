```python
import tkinter as tk
import math

radius = 200
centr = {'x':300, 'y':300}
angle = 0
speed = 0.01
direction = 1 # управление направлением движения

def draw(canvas):
    canvas.create_oval(
        centr['x'] - radius, centr['y'] - radius,
        centr['y'] + radius, centr['y'] + radius
    )

def move(canvas, point, angle):
    x = centr['x'] + radius * math.cos(angle)
    y = centr['y'] + radius * math.sin(angle)

    canvas.coords(point, x - 5, y - 5, x + 5, y + 5)

    root.after(10, move, canvas, point, angle + speed)
speed = speed*direction
root = tk.Tk()
root.geometry("600x600")
canvas = tk.Canvas(root, width=600, height=600)
canvas.pack()
draw(canvas)
point = canvas.create_oval(centr['x'], centr['y'], centr['x'], centr['y'], fill="yellow", outline="blue", width=3)
move(canvas, point, angle)
root.mainloop()
```
